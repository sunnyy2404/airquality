```python
data=data.drop(['Toluene','O3','Xylene','NH3','Benzene','month','year'],axis=1)
```

```python
import numpy as np
median = data['AQI_Bucket'].median()
mean = np.mean(data['AQI_Bucket'])

# Calculate the MAD
mad = data['AQI_Bucket'].mad()
```

```python
LC=median-(3*mad)
UC=median+(3*mad)
```

```python
LC,UC
```

```python

data1=data[(data.AQI_Bucket>LC)|(data.AQI_Bucket<UC)]
data1
```

```python

x=data1[['PM2.5','PM10','NO','NO2','NOx','CO','SO2']]
y=data1['AQI_Bucket']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn import metrics
from sklearn import preprocessing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict, train_test_split

regressor=DecisionTreeRegressor(max_depth=6, max_leaf_nodes=100)
regressor.fit(x_train, y_train)
```

```python
ac1=regressor.score(x_train,y_train)
ac1
```

```python
y_pred=regressor.predict(x_test)
```

```python
import numpy as numpy
np.sqrt(np.mean(np.square(y_test-y_pred)))
```

```python
np.mean(abs(y_test-y_pred)) 
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
print('RMSE value:',np.sqrt(mean_squared_error(y_test, y_pred)))
print('Mean absolute error:',mean_absolute_error(y_test, y_pred))
print('R2 Score:',r2_score(y_test, y_pred))
```
